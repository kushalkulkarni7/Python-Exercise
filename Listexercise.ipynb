```python
#second largest number in list

#function
def second_largest(list):
    list.sort()
    return list[-2]

#input of list
li=[]
n=int(input("Enter size of list "))
for i in range(0,n):
    e=int(input("Enter element of list "))
    li.append(e)

#smallest
print("second largest in ",li,"is")
print(second_largest(li))
```

Q.7 Write a Python program to find the common elements between two lists.

```python
# Input lists
a=[2,3,4,5]
b=[3,5,7,9]
def common(a,b):
  c = [value for value in a if value in b]
  return c
d=common(a,b)
print(d)
```

Q.10 Write a Python program to remove the elements of a list that are divisible by 3.
