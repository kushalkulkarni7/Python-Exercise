Q.8 Write a Python program to find the most frequent character in a string.

```python
# Input string
input_str = input("Enter a string: ")

# Initialize a dictionary to store character frequencies
char_frequency = {}

# Iterate through the string and count the frequency of each character
for char in input_str:
    # Increment the frequency count for the current character
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

# Find the character with the highest frequency
most_frequent_char = None
max_frequency = 0

for char, frequency in char_frequency.items():
    if frequency > max_frequency:
        max_frequency = frequency
        most_frequent_char = char

# Print the most frequent character and its frequency
if most_frequent_char is not None:
    print(f"The most frequent character is '{most_frequent_char}' with a frequency of {max_frequency}.")
else:
    print("The string is empty, there are no characters to count.")
```

Q.9 Write a Python program to check if a string is a valid email address.

```python
#re is module for regular expression
import re

# Input email address
email = input("Enter an email address: ")

# Regular expression pattern for a simple email validation
pattern = r'^[\w\.-]+@[\w\.-]+$'

# Use the re.match function to check if the email matches the pattern
if re.match(pattern, email):
    print(" It is a valid email address.")
else:
    print("It  is not a valid email address.")
```
