2. Write a Python program to concatenate two tuples.

3. Write a Python program to convert a tuple to a list.

```python
tuple = (20,30,50,70)
print("Tuple is : ",tuple)
# convert tuple into list
lst = list(tuple)
print("Tuple is converted into list is as: ",lst)
```

5. Write a Python program to check if an element exists in a tuple.

```python
# Input tuple
input_tuple = tuple(input("Enter elements of the tuple separated by spaces: ").split())

# Element to check
element_to_check = input("Enter the element to check for: ")

# Check if the element exists in the tuple
if element_to_check in input_tuple:
    print(f"{element_to_check} exists in the tuple.")
else:
    print(f"{element_to_check} does not exist in the tuple.")
```
